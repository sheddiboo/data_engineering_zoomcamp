```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark.context import SparkContext
```

```python
conf = SparkConf() \
    .setMaster('local[*]') \
    .setAppName('s3_taxi_data') \
    .set("spark.jars.packages", "org.apache.hadoop:hadoop-aws:3.3.2,com.amazonaws:aws-java-sdk-bundle:1.11.1026") \
    .set("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .set("spark.hadoop.fs.s3a.aws.credentials.provider", "com.amazonaws.auth.DefaultAWSCredentialsProviderChain") \
    .set("spark.hadoop.fs.s3a.endpoint", "s3.eu-west-1.amazonaws.com")
```

```python
sc = SparkContext.getOrCreate(conf=conf)

spark = SparkSession.builder \
    .config(conf=sc.getConf()) \
    .getOrCreate()
```

```python
df_green = spark.read.csv(
    's3a://zoomcamp-data-shedrack-2026/raw/greendata/*', 
    header=True, 
    inferSchema=True
)
```

```python
print(f"Total rows: {df_green.count():,}")
df_green.show(5)
```
